```python
sns.set_palette('bright')

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 6))

# Список признаков
features = [room1['district'], room2['district'], room3['district'], room4_5['district']]

# Построение гистограмм для каждого признака
for i, ax in enumerate(axes.flatten()):
    feature = features[i]
    values = feature.value_counts().head(5)  
    
    # Построение гистограммы
    ax.bar(values.index, values.values)
    ax.set_title('Районы с наибольшим количеством объявлений')
    ax.set_xlabel('Район')
    ax.set_ylabel('Количество объявлений')


# Размещение подграфиков
plt.tight_layout()

plt.rcParams.update({'font.size': 8})
plt.show()
```
